# Masking Strategy Comparison for PatchTST Self-Supervised Learning

This notebook evaluates the following masking strategies for PatchTST's masked patch reconstruction pretraining on the ETTh1 dataset:
1. Zero masking (baseline): Replace masked patches with zeros
2. Gaussian noise: Add Gaussian noise to masked patches

## Experimental Setup
- Masking ratio: 40%
- Patching configs: `patch_len=12, stride=12` (non-overlapping)
- Pre-training: 1 seed (2021), 10 epochs
- Fine-tuning: 5 seeds (2019-2023)
- Forecast horizons: {96, 192, 336, 720}

## 3. Gaussian Noise Addition

Unlike zero masking which completely removes information, Gaussian noise masking adds stochastic perturbations to the original signal:
$x_\text{masked} = x_\text{original} + \epsilon$, where $\epsilon \sim N(0, \sigma^2)$.

We evaluate three noise levels σ ∈ {0.1, 0.3, 0.5} to determine optimal magnitude. All candidates are tested with 10 pre-training epochs across 5 seeds on horizon 96.
