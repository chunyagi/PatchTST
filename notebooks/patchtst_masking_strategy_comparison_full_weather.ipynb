## 3. Gaussian Noise Addition

Unlike zero masking which completely removes information, Gaussian noise masking adds stochastic perturbations to the original signal:
$x_\text{masked} = x_\text{original} + \epsilon$, where $\epsilon \sim N(0, \sigma^2)$.

We evaluate three noise levels σ ∈ {0.03, 0.1, 0.3} to determine optimal magnitude. All candidates are tested with 100 pretraining epochs using seed 2021 on horizon 96 (only one seed to save compute).

**Note:** The search grid is centered around σ = 0.1 based on ETTh1 results.



```python
# Sigma 0.03
!python patchtst_pretrain.py --dset weather --batch_size 256 --d_ff 256 --use_gaussian_noise 1 --noise_std 0.03 --n_epochs_pretrain 100 --patient 7 --pretrained_model_id 3 --seed 2021
!python patchtst_finetune.py --dset weather --batch_size 256 --pretrained_model ./saved_models/weather/masked_patchtst/based_model/patchtst_pretrained_cw512_patch12_stride12_epochs-pretrain100_mask0.4_gnoise0.03_model3.pth --is_finetune 1 --finetuned_model_id 3 --seed 2021
```

```python
# Sigma 0.1
!python patchtst_pretrain.py --dset weather --batch_size 256 --d_ff 256 --use_gaussian_noise 1 --noise_std 0.1 --n_epochs_pretrain 100 --patient 7 --pretrained_model_id 3 --seed 2021
!python patchtst_finetune.py --dset weather --batch_size 256 --pretrained_model ./saved_models/weather/masked_patchtst/based_model/patchtst_pretrained_cw512_patch12_stride12_epochs-pretrain100_mask0.4_gnoise0.1_model3.pth --is_finetune 1 --finetuned_model_id 3 --seed 2021
```
